```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Customer_Churn.csv')
```

```python
df
```

```python
df["TotalCharges"] = df['TotalCharges'].replace(' ',"0")
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

```python
#converted 0 and 1 value in senior citizen column to yes/no.
```

```python
def conv(value):
  if value == 1:
    return 'Yes'
  else:
    return 'No'
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
ax = sns.countplot(x=df['Churn'])
ax.bar_label(ax.containers[0])
plt.title('count of customers by churn')
```

```python
gb = df['Churn'].value_counts()
plt.figure(figsize=(3,3))
plt.pie(gb,labels=gb.index,autopct='%1.2f%%')
plt.title('percentage of churned customers')
plt.show()
```

```python
#from the given pie chart we can conclude that 26.54% of our customers have been churned out.
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x='gender',data = df,hue = 'Churn')
plt.show()
```

```python
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
plt.figure(figsize=(3,3))
sns.barplot(x='SeniorCitizen', y='Count', hue='Churn',data=grouped)
plt.title('Churn by Senior Citizen')
plt.ylabel('Number of Customers')
plt.show()
```

```python
#Comparative a greater percentage of people in senior citizen have churned.
```

```python
plt.figure(figsize=(5,3))
sns.histplot(x='tenure',data = df,hue = 'Churn',bins=50)
plt.title('tenure distribution by churn')
plt.show()
```

```python
#people who have used our services for a long time have stayed while people who used our services for first few months
# have been churned.
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x='Contract',data=df,hue='Churn')
ax.bar_label(ax.containers[0])
plt.title('count of customers by contract')
plt.show()
```

```python
#people who have month to month contract are more likely to churn as compared to those who have yearly contract.
```

```python
# List of columns to plot
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]
# Define grid size (e.g., 3 rows Ã— 3 columns)
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols
# Set figure size
plt.figure(figsize=(12, 12))
# Loop through columns and plot
for i, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(data=df, x=col, order=df[col].value_counts().index,hue='Churn',legend=False)
    plt.title(col)
    plt.xticks(rotation=45)
    plt.tight_layout()
plt.suptitle('Count Plots for Customer Services', y=1.02, fontsize=16)
plt.show()
```

```python
#Customers who use support and protection services like **TechSupport**,
#**OnlineSecurity**, and **OnlineBackup** show **significantly lower churn**,
#indicating these features enhance retention. In contrast, **Fiber optic internet users**
#exhibit **higher churn**, suggesting possible service or pricing issues.
# **Streaming services** like TV and movies do not strongly impact churn.
#Encouraging adoption of support services and improving fiber customer experience could help reduce overall churn.
```
